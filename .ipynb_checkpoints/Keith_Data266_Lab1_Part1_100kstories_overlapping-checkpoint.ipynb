```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from datasets import load_dataset
import matplotlib.pyplot as plt
import math
import random
import numpy as np
import os
import shutil

# Reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')
```

## 1.1 Data Preprocessing

```python
from bisect import bisect_right

SEQ_LEN    = 128
BATCH_SIZE = 128

dataset = load_dataset('roneneldan/TinyStories', split='train')

train_stories = dataset.select(range(100_000))['text']
val_stories   = dataset.select(range(100_000, 110_000))['text']

# Build vocab from all text
all_text_for_vocab = ''.join(train_stories) + ''.join(val_stories)
chars       = sorted(set(all_text_for_vocab))
vocab_size  = len(chars)
char_to_idx = {ch: i for i, ch in enumerate(chars)}
idx_to_char = {i: ch for ch, i in char_to_idx.items()}

encode = lambda s: [char_to_idx[c] for c in s if c in char_to_idx]
decode = lambda l: ''.join([idx_to_char[i] for i in l])

print(f'Train stories: 100,000')
print(f'Val stories:   10,000')
print(f'Vocab size:    {vocab_size}')


class StoryCharDataset(Dataset):
    """Sliding-window dataset that respects story boundaries."""
    def __init__(self, stories, char_to_idx, seq_len=128):
        self.seq_len = seq_len
        self.encoded_stories  = []
        self.windows_per_story = []

        for story in stories:
            ids = [char_to_idx[c] for c in story if c in char_to_idx]
            if len(ids) > seq_len:
                self.encoded_stories.append(torch.tensor(ids, dtype=torch.long))
                self.windows_per_story.append(len(ids) - seq_len)

        self.cumulative = []
        total = 0
        for w in self.windows_per_story:
            total += w
            self.cumulative.append(total)

    def __len__(self):
        return self.cumulative[-1] if self.cumulative else 0

    def __getitem__(self, idx):
        story_idx  = bisect_right(self.cumulative, idx)
        prev_total = 0 if story_idx == 0 else self.cumulative[story_idx - 1]
        local_idx  = idx - prev_total
        story_ids  = self.encoded_stories[story_idx]
        x = story_ids[local_idx     : local_idx + self.seq_len]
        y = story_ids[local_idx + 1 : local_idx + self.seq_len + 1]
        return x, y


train_dataset = StoryCharDataset(train_stories, char_to_idx, SEQ_LEN)
val_dataset   = StoryCharDataset(val_stories,   char_to_idx, SEQ_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,  drop_last=True)
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE, shuffle=False, drop_last=False)

print(f'Train windows: {len(train_dataset):,}')
print(f'Val windows:   {len(val_dataset):,}')
print(f'Train batches: {len(train_loader):,}')
print(f'Val batches:   {len(val_loader):,}')
```

## 1.2 GPT Model Implementation

```python
class MultiHeadSelfAttention(nn.Module):
    def __init__(self, embed_dim, num_heads):
        super().__init__()
        assert embed_dim % num_heads == 0
        self.num_heads = num_heads
        self.head_dim  = embed_dim // num_heads

        self.q_proj    = nn.Linear(embed_dim, embed_dim, bias=False)
        self.k_proj    = nn.Linear(embed_dim, embed_dim, bias=False)
        self.v_proj    = nn.Linear(embed_dim, embed_dim, bias=False)
        self.out_proj  = nn.Linear(embed_dim, embed_dim)
        self.attn_drop = nn.Dropout(0.1)

    def forward(self, x):
        B, T, C = x.shape
        H, D = self.num_heads, self.head_dim

        q = self.q_proj(x).view(B, T, H, D).transpose(1, 2)  # (B,H,T,D)
        k = self.k_proj(x).view(B, T, H, D).transpose(1, 2)
        v = self.v_proj(x).view(B, T, H, D).transpose(1, 2)

        scores = (q @ k.transpose(-2, -1)) / math.sqrt(D)    # (B,H,T,T)

        # Causal mask: upper triangle set to -inf
        mask = torch.triu(torch.ones(T, T, device=x.device), diagonal=1).bool()
        scores = scores.masked_fill(mask, float('-inf'))

        attn = F.softmax(scores, dim=-1)
        attn = self.attn_drop(attn)
        out  = (attn @ v).transpose(1, 2).contiguous().view(B, T, C)
        return self.out_proj(out)


class FeedForward(nn.Module):
    def __init__(self, embed_dim, ff_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(embed_dim, ff_dim),
            nn.GELU(),
            nn.Linear(ff_dim, embed_dim),
            nn.Dropout(0.1),
        )

    def forward(self, x):
        return self.net(x)


class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.ln1  = nn.LayerNorm(embed_dim)
        self.attn = MultiHeadSelfAttention(embed_dim, num_heads)
        self.ln2  = nn.LayerNorm(embed_dim)
        self.ff   = FeedForward(embed_dim, ff_dim)

    def forward(self, x):
        x = x + self.attn(self.ln1(x))  # residual
        x = x + self.ff(self.ln2(x))    # residual
        return x


class GPTModel(nn.Module):
    def __init__(self, vocab_size, embed_dim, num_heads, ff_dim, num_layers, block_size):
        super().__init__()
        self.token_emb = nn.Embedding(vocab_size, embed_dim)       # learnable token embeddings
        self.pos_emb   = nn.Embedding(block_size, embed_dim)       # learnable positional embeddings
        self.drop      = nn.Dropout(0.1)
        self.blocks    = nn.Sequential(*[
            TransformerBlock(embed_dim, num_heads, ff_dim) for _ in range(num_layers)
        ])
        self.ln_f    = nn.LayerNorm(embed_dim)
        self.lm_head = nn.Linear(embed_dim, vocab_size, bias=False)  # language modelling head

        # Weight tying between token embedding and LM head
        self.token_emb.weight = self.lm_head.weight

        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.normal_(m.weight, mean=0.0, std=0.02)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)
            elif isinstance(m, nn.Embedding):
                nn.init.normal_(m.weight, mean=0.0, std=0.02)

    def forward(self, idx, targets=None):
        B, T = idx.shape
        positions = torch.arange(T, device=idx.device)

        x = self.drop(self.token_emb(idx) + self.pos_emb(positions))
        x = self.blocks(x)
        x = self.ln_f(x)
        logits = self.lm_head(x)  # (B, T, vocab_size)

        loss = None
        if targets is not None:
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), targets.view(-1))
        return logits, loss

    @torch.no_grad()
    def generate(self, idx, max_new_tokens, temperature=1.0):
        self.eval()
        for _ in range(max_new_tokens):
            idx_cond = idx[:, -SEQ_LEN:]
            logits, _ = self(idx_cond)
            logits = logits[:, -1, :]  # (B, vocab)
            if temperature is None or temperature <= 1e-8:
                next_id = torch.argmax(logits, dim=-1, keepdim=True)
            else:
                logits  = logits / temperature
                probs   = F.softmax(logits, dim=-1)
                next_id = torch.multinomial(probs, num_samples=1)
            idx = torch.cat([idx, next_id], dim=1)
        return idx
```

```python
# Hyperparameters
EMBED_DIM    = 256
NUM_HEADS    = 8
FF_DIM       = 1024
NUM_LAYERS   = 6
EPOCHS       = 10
BASE_LR      = 3e-3
WARMUP_STEPS = 200

model = GPTModel(
    vocab_size=vocab_size,
    embed_dim=EMBED_DIM,
    num_heads=NUM_HEADS,
    ff_dim=FF_DIM,
    num_layers=NUM_LAYERS,
    block_size=SEQ_LEN,
).to(device)

total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f'Trainable parameters: {total_params:,}')
```

## 1.3 Training and Text Generation

```python
optimizer   = torch.optim.AdamW(model.parameters(), lr=BASE_LR, weight_decay=1e-2)
total_steps = EPOCHS * len(train_loader)

def lr_lambda(step):
    if step < WARMUP_STEPS:
        return step / max(1, WARMUP_STEPS)                        # linear warm-up
    progress = (step - WARMUP_STEPS) / max(1, total_steps - WARMUP_STEPS)
    return max(0.1, 0.5 * (1 + math.cos(math.pi * progress)))    # cosine decay

scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)
```

```python
@torch.no_grad()
def evaluate(loader, max_batches=30):
    model.eval()
    total_loss = 0
    for i, (x, y) in enumerate(loader):
        if i >= max_batches:
            break
        x, y = x.to(device), y.to(device)
        _, loss = model(x, y)
        total_loss += loss.item()
    return total_loss / min(max_batches, len(loader))


out_dir = '/content/part1_output'
os.makedirs(out_dir, exist_ok=True)

train_losses, val_losses = [], []
global_step  = 0
best_val_loss = float('inf')

for epoch in range(1, EPOCHS + 1):
    model.train()
    epoch_loss = 0

    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        _, loss = model(x, y)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        epoch_loss += loss.item()
        global_step += 1

    avg_train = epoch_loss / len(train_loader)
    avg_val   = evaluate(val_loader)
    train_losses.append(avg_train)
    val_losses.append(avg_val)
    current_lr = scheduler.get_last_lr()[0]
    print(f'Epoch {epoch:02d}/{EPOCHS} | Train Loss: {avg_train:.4f} | Val Loss: {avg_val:.4f} | LR: {current_lr:.6f}')

    # Save checkpoint every epoch
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'train_loss': avg_train,
        'val_loss': avg_val,
        'char_to_idx': char_to_idx,
        'idx_to_char': idx_to_char,
        'vocab_size': vocab_size,
        'block_size': SEQ_LEN,
    }, os.path.join(out_dir, f'checkpoint_epoch{epoch:02d}.pt'))

    # Save best model separately
    if avg_val < best_val_loss:
        best_val_loss = avg_val
        torch.save(model.state_dict(), os.path.join(out_dir, 'model.pth'))
        print(f'  ✓ Best model saved (val loss: {best_val_loss:.4f})')
```

```python
# Plot and save training / validation loss curves
plt.figure(figsize=(9, 4))
plt.plot(range(1, EPOCHS + 1), train_losses, marker='o', label='Train Loss')
plt.plot(range(1, EPOCHS + 1), val_losses,   marker='s', label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Cross-Entropy Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('/content/loss_curves.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'Final Train Loss: {train_losses[-1]:.4f} | Final Val Loss: {val_losses[-1]:.4f}')
```

```python
# Save final checkpoint with full metadata
torch.save({
    'model_state_dict': model.state_dict(),
    'char_to_idx': char_to_idx,
    'idx_to_char': idx_to_char,
    'vocab_size': vocab_size,
    'block_size': SEQ_LEN,
}, os.path.join(out_dir, 'tinygpt_checkpoint.pt'))

print('Saved to:', out_dir)
print('Files:', os.listdir(out_dir))
```

```python
# Text generation
def generate_text(prompt, max_new_tokens=200, temperature=1.0):
    model.eval()
    encoded = encode(prompt)
    context = torch.tensor(encoded, dtype=torch.long, device=device).unsqueeze(0)
    out = model.generate(context, max_new_tokens=max_new_tokens, temperature=temperature)
    return decode(out[0].tolist())


prompt = 'Once upon a time'

print('--- Greedy Decoding (temperature=0) ---')
print(generate_text(prompt, max_new_tokens=200, temperature=0))

print('\n--- Temperature Sampling (T=0.8) ---')
print(generate_text(prompt, max_new_tokens=200, temperature=0.8))

print('\n--- Temperature Sampling (T=1.2) ---')
print(generate_text(prompt, max_new_tokens=200, temperature=1.2))
```

## 1.4 Sequence Model Failure Analysis

```python
failure_samples = []
prompts = ['Once upon a time', 'The little dog', 'One day a girl named']

for p in prompts:
    out = generate_text(p, max_new_tokens=300, temperature=1.0)
    failure_samples.append(out)
    print(f'Prompt: "{p}"')
    print(out)
    print('-' * 70)
```

```python
failure_cases = [
    {
        'snippet': '*** PASTE SNIPPET 1 HERE ***',
        'type': 'Repetition',
        'observation': 'The model repeatedly generates the same phrase or word sequence. '
                       'Greedy decoding can collapse into a high-probability loop once a '
                       'likely continuation is found, as there is no penalty for repeated tokens.',
    },
    {
        'snippet': '*** PASTE SNIPPET 2 HERE ***',
        'type': 'Loss of Coherence',
        'observation': 'The narrative topic shifts abruptly mid-sentence. The model loses '
                       'track of the story context over longer sequences, producing text that '
                       'is grammatically plausible locally but globally incoherent.',
    },
    {
        'snippet': '*** PASTE SNIPPET 3 HERE ***',
        'type': 'Hallucination / Nonsense',
        'observation': 'High temperature sampling causes the model to select low-probability '
                       'tokens, producing nonsensical phrases. The model also loses long-range '
                       'coherence, restarting the story mid-generation rather than continuing '
                       'the same narrative thread.',
    },
]

for i, fc in enumerate(failure_cases, 1):
    print(f'Failure Case {i}')
    print(f'  Type       : {fc["type"]}')
    print(f'  Snippet    : ...{fc["snippet"]}...')
    print(f'  Observation: {fc["observation"]}')
    print()
```
